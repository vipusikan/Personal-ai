```python
import pandas as pd
import random

random.seed(42)

topics = [
    "SDLC","Agile Model","Scrum","UML",
    "Use Case Diagram","Class Diagram",
    "Sequence Diagram","Software Testing",
    "Unit Testing","Integration Testing",
    "System Testing","Version Control",
    "Software Maintenance"
]

# More templates = more rows
explain_templates = [
    "Explain {}",
    "Explain {} in simple words",
    "Explain {} briefly",
    "Explain {} with example",
    "What is {}?",
    "Define {}",
    "Meaning of {}",
    "Tell me about {}",
    "I want to learn {}",
    "Teach me {}",
    "Describe {}",
    "Can you explain {}?",
    "Give notes on {}",
    "Summarize {}",
    "Give advantages and disadvantages of {}",
    "Why do we use {}?",
    "When to use {}?",
    "How {} works?",
    "Steps of {}",
    "Explain {} for exam"
]

quiz_templates = [
    "Quiz me on {}",
    "Give me a quiz on {}",
    "Test me on {}",
    "Give MCQ on {}",
    "MCQ questions on {}",
    "Questions for {}",
    "Practice questions on {}",
    "Start quiz on {}",
    "Give 5 questions on {}",
    "Give 10 questions on {}",
    "Give easy questions on {}",
    "Give hard questions on {}",
    "I want quiz for {}"
]

recommend_templates = [
    "Recommend resources for {}",
    "Give YouTube videos for {}",
    "Give websites for {}",
    "Best learning resources for {}",
    "Suggest tutorials for {}",
    "Any links for {}?",
    "Where can I learn {}?",
    "Best book for {}",
    "Recommend W3School for {}",
    "Give blogs for {}"
]

progress_templates = [
    "Show my progress",
    "How is my progress?",
    "My performance",
    "Show my mastery",
    "How am I doing?",
    "Show my weak topics",
    "What is my score?",
    "Show my quiz history",
    "Progress report",
    "My learning report"
]

check_answer_templates = [
    "Check my answer",
    "Is my answer correct?",
    "Evaluate my answer",
    "Can you check this answer?",
    "Is this correct?",
    "Correct or wrong?",
    "Mark my answer",
    "Give feedback for my answer"
]

greetings = [
    "hi","hello","hey","hii","hiii","yo","hai",
    "good morning","good afternoon","good evening",
    "whats up","what's up","sup","how are you"
]

thanks = [
    "thanks","thank you","ok thanks","cool thanks",
    "great thanks","nice","good","awesome","thank u","ty"
]

# VERY IMPORTANT: add lots of noise/short texts
other_noise = [
    "h","hh","h.","a","x","??","...","ok","k","hmm","hahaha",
    "asdf","random","what","help","lol","?","??!!",".....",
    "hey?","hello?","yo?","hiiii","okey","okk","idk","h?"
]

# Variation maker (no ML, just data augmentation)
def variations(s):
    s = s.strip()
    return list(set([
        s,
        s.lower(),
        s.upper(),
        s.capitalize(),
        s + "?",
        s + "!",
        "pls " + s,
        "please " + s,
        s.replace(" ", ""),
    ]))

rows = []

# greetings (with variations)
for g in greetings:
    for v in variations(g):
        rows.append([v, "greeting"])

# thanks
for t in thanks:
    for v in variations(t):
        rows.append([v, "thanks"])

# noise/other
for n in other_noise:
    for v in variations(n):
        rows.append([v, "other"])

# progress
for p in progress_templates:
    for v in variations(p):
        rows.append([v, "progress"])

# check_answer
for c in check_answer_templates:
    for v in variations(c):
        rows.append([v, "check_answer"])

# explain/quiz/recommend with topics + variations
for topic in topics:
    for temp in explain_templates:
        for v in variations(temp.format(topic)):
            rows.append([v, "explain"])
    for temp in quiz_templates:
        for v in variations(temp.format(topic)):
            rows.append([v, "quiz"])
    for temp in recommend_templates:
        for v in variations(temp.format(topic)):
            rows.append([v, "recommend"])

df = pd.DataFrame(rows, columns=["text", "intent"]).drop_duplicates()
df["text"] = df["text"].astype(str).str.strip()

df.to_csv("intent_dataset_expanded.csv", index=False)

print("✅ Saved: intent_dataset_expanded.csv")
print("Total rows:", len(df))
df["intent"].value_counts()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import joblib
from pathlib import Path

df = pd.read_csv("intent_dataset_expanded.csv")

X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["intent"],
    test_size=0.2,
    random_state=42,
    stratify=df["intent"]
)

model = Pipeline([
    ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=1)),
    ("clf", LinearSVC(class_weight="balanced"))
])

model.fit(X_train, y_train)
pred = model.predict(X_test)

print("=== INTENT MODEL REPORT ===")
print(classification_report(y_test, pred))

# save model (joblib)
Path("../backend/models").mkdir(parents=True, exist_ok=True)
joblib.dump(model, "../backend/models/intent_model.joblib")
print("✅ Saved: ../backend/models/intent_model.joblib")
```

```python
import pandas as pd

df = pd.read_csv("intent_dataset_expanded.csv")

# Find smallest class size
min_count = df["intent"].value_counts().min()
print("Smallest class size:", min_count)

# Instead of using smallest (64), let's cap large classes to 1000
MAX_SAMPLES = 1000

balanced_rows = []

for intent, group in df.groupby("intent"):
    if len(group) > MAX_SAMPLES:
        balanced_rows.append(group.sample(MAX_SAMPLES, random_state=42))
    else:
        balanced_rows.append(group)

balanced_df = pd.concat(balanced_rows)

print(balanced_df["intent"].value_counts())

balanced_df.to_csv("intent_dataset_balanced.csv", index=False)
print("✅ Saved: intent_dataset_balanced.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import pandas as pd
import joblib
from pathlib import Path

df = pd.read_csv("intent_dataset_balanced.csv")

X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["intent"],
    test_size=0.2,
    random_state=42,
    stratify=df["intent"]
)

model = Pipeline([
    ("tfidf", TfidfVectorizer(
        ngram_range=(1,2),
        min_df=2,
        sublinear_tf=True,
        strip_accents="unicode"
    )),
    ("clf", LinearSVC(
        class_weight="balanced",
        C=0.8
    ))
])

model.fit(X_train, y_train)
pred = model.predict(X_test)

print("=== IMPROVED INTENT MODEL REPORT ===")
print(classification_report(y_test, pred))

joblib.dump(model, "../backend/models/intent_model.joblib")
print("✅ Final improved model saved.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib

# Load your uploaded dataset
df = pd.read_csv("Software Questions.csv", encoding="latin1")

# Remove empty rows
df = df.dropna(subset=["Question", "Answer"])

questions = df["Question"].astype(str).tolist()
answers = df["Answer"].astype(str).tolist()

# Create TF-IDF model
vectorizer = TfidfVectorizer(
    stop_words="english",
    ngram_range=(1,2),
    min_df=2,
    sublinear_tf=True
)

vectorizer.fit(answers)

# Save vectorizer
joblib.dump(vectorizer, "../backend/models/descriptive_vectorizer.joblib")

print("✅ Descriptive scoring model saved")
print("Total theory questions:", len(questions))
```
