```python
import json
import joblib
from pathlib import Path

# ensure models folder exists
Path("../backend/models").mkdir(parents=True, exist_ok=True)

def export_linear_svc_pipeline(joblib_path, out_json_path):
    pipe = joblib.load(joblib_path)

    tfidf = pipe.named_steps["tfidf"]
    clf = pipe.named_steps["clf"]

    payload = {
        "ngram_range": list(tfidf.ngram_range),
        "lowercase": tfidf.lowercase,
        "vocabulary_": tfidf.vocabulary_,
        "idf_": tfidf.idf_.tolist(),
        "classes_": clf.classes_.tolist(),
        "coef_": clf.coef_.tolist(),
        "intercept_": clf.intercept_.tolist()
    }

    with open(out_json_path, "w", encoding="utf-8") as f:
        json.dump(payload, f)

# export intent model
export_linear_svc_pipeline(
    "../backend/models/intent_model.joblib",
    "../backend/models/intentModel.json"
)

# export topic model
export_linear_svc_pipeline(
    "../backend/models/topic_model.joblib",
    "../backend/models/topicModel.json"
)

print("Export complete.")
```

```python
import pandas as pd
from pathlib import Path
import random

random.seed(42)

TOPICS = [
  "SDLC",
  "Agile Model",
  "Scrum",
  "UML",
  "Use Case Diagram",
  "Class Diagram",
  "Sequence Diagram",
  "Software Testing",
  "Unit Testing",
  "Integration Testing",
  "System Testing",
  "Version Control",
  "Software Maintenance"
]

# ---- Base facts per topic (used to auto-create variants) ----
FACTS = {
 "SDLC": {
   "defs": ["Software Development Life Cycle", "requirements, design, implementation, testing, deployment, maintenance"],
   "keywords": ["requirements", "design", "implementation", "testing", "deployment", "maintenance"],
   "models": ["Waterfall", "Spiral", "Incremental", "Agile"]
 },
 "Agile Model": {
   "defs": ["iterative and incremental", "frequent delivery", "responding to change"],
   "keywords": ["backlog", "sprint", "standup", "review", "retrospective", "increment"],
   "models": ["Scrum", "Kanban", "XP"]
 },
 "Scrum": {
   "defs": ["framework with sprints", "roles: Product Owner, Scrum Master, Developers"],
   "keywords": ["Product Backlog", "Sprint Backlog", "Increment", "Sprint Planning", "Daily Scrum", "Review", "Retrospective"],
   "models": ["Sprint", "Backlog", "DoD"]
 },
 "UML": {
   "defs": ["Unified Modeling Language", "standard modeling language"],
   "keywords": ["structural", "behavioral", "diagrams", "model"],
   "models": ["Class", "Sequence", "Use Case", "Activity"]
 },
 "Use Case Diagram": {
   "defs": ["actors and use cases", "functional requirements"],
   "keywords": ["actor", "use case", "include", "extend", "system boundary"],
   "models": ["include", "extend", "generalization"]
 },
 "Class Diagram": {
   "defs": ["classes and relationships", "attributes and methods"],
   "keywords": ["association", "inheritance", "aggregation", "composition", "multiplicity"],
   "models": ["association", "generalization", "composition"]
 },
 "Sequence Diagram": {
   "defs": ["message flow over time", "object interactions"],
   "keywords": ["lifeline", "activation", "message", "alt", "loop", "return"],
   "models": ["sync", "async", "return"]
 },
 "Software Testing": {
   "defs": ["find defects", "verify requirements"],
   "keywords": ["functional", "non-functional", "regression", "test case", "defect"],
   "models": ["unit", "integration", "system", "acceptance"]
 },
 "Unit Testing": {
   "defs": ["tests smallest units", "fast and isolated"],
   "keywords": ["mock", "stub", "AAA", "TDD", "coverage"],
   "models": ["Arrange-Act-Assert", "TDD"]
 },
 "Integration Testing": {
   "defs": ["tests module interactions", "interface defects"],
   "keywords": ["top-down", "bottom-up", "big-bang", "API", "contract"],
   "models": ["top-down", "bottom-up", "big-bang"]
 },
 "System Testing": {
   "defs": ["tests whole system", "end-to-end"],
   "keywords": ["performance", "security", "compatibility", "recovery", "load"],
   "models": ["end-to-end", "UAT"]
 },
 "Version Control": {
   "defs": ["track changes", "collaboration with Git"],
   "keywords": ["commit", "branch", "merge", "pull request", "conflict", "tag"],
   "models": ["Git Flow", "trunk-based"]
 },
 "Software Maintenance": {
   "defs": ["post-deployment work", "fixes and enhancements"],
   "keywords": ["corrective", "adaptive", "perfective", "preventive", "refactoring", "monitoring"],
   "models": ["corrective", "adaptive", "perfective", "preventive"]
 }
}

def make_options(correct, wrongs):
    # ensure 3 wrong options + 1 correct
    options = wrongs[:3] + [correct]
    random.shuffle(options)
    letters = ["A","B","C","D"]
    mapping = dict(zip(letters, options))
    answer = [k for k,v in mapping.items() if v == correct][0]
    return mapping, answer

def gen_question(topic, i):
    f = FACTS[topic]
    diff = random.choice([1,1,2,2,3])  # more easy/medium

    # patterns
    patterns = [
      ("definition", lambda: (f"What best describes {topic}?", f["defs"][0])),
      ("keyword",    lambda: (f"Which term is most related to {topic}?", random.choice(f["keywords"]))),
      ("true_purpose",lambda: (f"The main purpose of {topic} is to:", random.choice(f["defs"]))),
      ("not_related", lambda: (f"Which is NOT related to {topic}?", "Not related")),
      ("concept",    lambda: (f"Which concept belongs to {topic}?", random.choice(f["models"]))),
    ]

    p, builder = random.choice(patterns)
    qtext, correct = builder()

    # wrong pool (generic wrong answers)
    wrong_pool = [
      "Bluetooth protocol", "CPU scheduling", "Database normalization only", "Network routing",
      "Cooking recipe", "Hardware installation", "Email SMTP", "DNS resolution",
      "Random guessing", "User interface theme"
    ]

    # add topic-specific wrongs
    topic_wrongs = []
    for t2 in TOPICS:
      if t2 != topic:
        topic_wrongs.append(t2)

    wrongs = wrong_pool + topic_wrongs
    random.shuffle(wrongs)

    if p == "not_related":
      # for NOT related, correct is "Not related"; make other 3 be related keywords/models
      related = list(set(f["keywords"] + f["models"]))
      random.shuffle(related)
      mapping, answer = make_options("Not related", related + wrongs)
      return qtext, mapping, answer, diff

    mapping, answer = make_options(correct, wrongs)
    return qtext, mapping, answer, diff

# ---- Build ~520 questions (40 per topic) ----
rows = []
qid = 1
PER_TOPIC = 40  # 40 * 13 = 520

for topic in TOPICS:
    for i in range(PER_TOPIC):
        qtext, mapping, ans, diff = gen_question(topic, i)
        rows.append({
            "id": qid,
            "topic": topic,
            "difficulty": diff,
            "question": qtext,
            "A": mapping["A"],
            "B": mapping["B"],
            "C": mapping["C"],
            "D": mapping["D"],
            "answer": ans
        })
        qid += 1

df_q = pd.DataFrame(rows)
Path("../backend/data").mkdir(parents=True, exist_ok=True)
df_q.to_csv("../backend/data/questions.csv", index=False)

print("✅ Saved ../backend/data/questions.csv")
print("Total questions:", len(df_q))
df_q.sample(5)
```

```python
import pandas as pd
from pathlib import Path
import random

random.seed(42)

topics = [
  "SDLC","Agile","Scrum","UML","Use case diagram","Class diagram","Sequence diagram",
  "software testing","unit testing","integration testing","system testing",
  "Git","version control","software maintenance"
]

greeting = [
 "hi", "hello", "hey", "good morning", "good evening", "hi tutor", "hello bot", "hey there"
]

explain = []
for t in topics:
    explain += [
      f"explain {t}",
      f"what is {t}?",
      f"give me notes on {t}",
      f"teach me {t}",
      f"define {t}",
      f"can you explain {t} with example?",
      f"explain {t} in simple words"
    ]

quiz = []
for t in topics:
    quiz += [
      f"quiz me on {t}",
      f"start a quiz on {t}",
      f"give mcq on {t}",
      f"ask questions about {t}",
      f"i want a quiz for {t}"
    ]

check_answer = []
for t in topics:
    check_answer += [
      f"check my answer on {t}",
      f"is my answer correct for {t}?",
      f"i chose option A, is it correct?",
      f"my answer is B, correct?",
      f"explain why option C is wrong"
    ]

recommend = []
for t in topics:
    recommend += [
      f"recommend resources for {t}",
      f"give me youtube links for {t}",
      f"best website to learn {t}",
      f"recommend tutorials on {t}",
      f"suggest materials for {t}"
    ]

other = [
 "tell me a joke", "what is your name", "open google", "who is the president", "play music",
 "can you do my assignment", "give me movie recommendations"
]

data = []
def add(intent, texts):
    for x in texts:
        data.append({"text": x, "intent": intent})

add("greeting", greeting)
add("explain", explain)
add("quiz", quiz)
add("check_answer", check_answer)
add("recommend", recommend)
add("other", other)

df_intent = pd.DataFrame(data).sample(frac=1, random_state=42).reset_index(drop=True)

Path("../data").mkdir(parents=True, exist_ok=True)
df_intent.to_csv("../data/intent_dataset.csv", index=False)

print("✅ Saved ../data/intent_dataset.csv")
print("Rows:", len(df_intent))
df_intent.head(10)
```

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import joblib

# Paths
intent_path = Path("../data/intent_dataset.csv")
Path("../backend/models").mkdir(parents=True, exist_ok=True)

# Load
df = pd.read_csv(intent_path)
df = df.dropna()

# Split
X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["intent"],
    test_size=0.2, random_state=42, stratify=df["intent"]
)

# Train model
intent_model = Pipeline([
    ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=1)),
    ("clf", LinearSVC())
])

intent_model.fit(X_train, y_train)
pred = intent_model.predict(X_test)

print("=== INTENT MODEL REPORT ===")
print(classification_report(y_test, pred, zero_division=0))

# Save joblib
joblib.dump(intent_model, "../backend/models/intent_model.joblib")
print("✅ Saved: ../backend/models/intent_model.joblib")
```

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import joblib

topic_path = Path("../data/topic_dataset.csv")
Path("../backend/models").mkdir(parents=True, exist_ok=True)

df = pd.read_csv(topic_path)
df = df.dropna()

X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["topic"],
    test_size=0.2, random_state=42, stratify=df["topic"]
)

topic_model = Pipeline([
    ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=1)),
    ("clf", LinearSVC())
])

topic_model.fit(X_train, y_train)
pred = topic_model.predict(X_test)

print("=== TOPIC MODEL REPORT ===")
print(classification_report(y_test, pred, zero_division=0))

joblib.dump(topic_model, "../backend/models/topic_model.joblib")
print("✅ Saved: ../backend/models/topic_model.joblib")
```

```python
import json
import joblib
from pathlib import Path

Path("../backend/models").mkdir(parents=True, exist_ok=True)

def export_linear_svc_pipeline(joblib_path, out_json_path):
    pipe = joblib.load(joblib_path)
    tfidf = pipe.named_steps["tfidf"]
    clf = pipe.named_steps["clf"]

    payload = {
        "ngram_range": list(tfidf.ngram_range),
        "lowercase": tfidf.lowercase,
        "vocabulary_": tfidf.vocabulary_,
        "idf_": tfidf.idf_.tolist(),
        "classes_": clf.classes_.tolist(),
        "coef_": clf.coef_.tolist(),
        "intercept_": clf.intercept_.tolist()
    }

    with open(out_json_path, "w", encoding="utf-8") as f:
        json.dump(payload, f)

export_linear_svc_pipeline("../backend/models/intent_model.joblib", "../backend/models/intentModel.json")
export_linear_svc_pipeline("../backend/models/topic_model.joblib", "../backend/models/topicModel.json")

print("✅ Exported:")
print(" - ../backend/models/intentModel.json")
print(" - ../backend/models/topicModel.json")
```
